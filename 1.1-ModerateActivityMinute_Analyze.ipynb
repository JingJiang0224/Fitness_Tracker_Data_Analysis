## Analyze moderate activity minute

```python
# Convert ActivityDate to datetime
dailyActivity_merged['ActivityDate'] = pd.to_datetime(dailyActivity_merged['ActivityDate'])

# Extract the year and week number from the ActivityDate
dailyActivity_merged['YearWeek'] = dailyActivity_merged['ActivityDate'].dt.strftime('%Y-%U')

# moderate activity minute
dailyActivity_merged['moderate_total_minute1'] = dailyActivity_merged['VeryActiveMinutes'] + dailyActivity_merged['FairlyActiveMinutes'] 
dailyActivity_merged['moderate_total_minute2'] = dailyActivity_merged['VeryActiveMinutes'] + dailyActivity_merged['FairlyActiveMinutes'] + dailyActivity_merged['LightlyActiveMinutes'] 
```

```python
# Group by Id and YearWeek, and calculate the total minutes per week for each user
total_minutes_per_week1 = dailyActivity_merged.groupby(['Id', 'YearWeek'])['moderate_total_minute1'].sum().reset_index()
total_minutes_per_week2 = dailyActivity_merged.groupby(['Id', 'YearWeek'])['moderate_total_minute2'].sum().reset_index()

# Group by Id to calculate the average minutes per week for each user
average_minutes_per_week1 = total_minutes_per_week1.groupby('Id')['moderate_total_minute1'].mean().reset_index()
average_minutes_per_week2 = total_minutes_per_week2.groupby('Id')['moderate_total_minute2'].mean().reset_index()

# Rename columns for clarity
average_minutes_per_week1.columns = ['Id', 'AverageTotalMinutesPerWeek1']
average_minutes_per_week2.columns = ['Id', 'AverageTotalMinutesPerWeek2']
```

```python
print(average_minutes_per_week1)
print(average_minutes_per_week2)
```

```python
average_minutes_per_week1.to_excel("average_minutes_per_week1.xlsx")
average_minutes_per_week2.to_excel("average_minutes_per_week2.xlsx")
```
