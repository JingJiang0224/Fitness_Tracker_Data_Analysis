```python
import pandas as pd
```

```python
hourlyCalories = pd.read_csv('hourlyCalories_merged.csv')
```

```python
hourlyCalories
```

### Duplicate data

```python
duplicates_specific = hourlyCalories.duplicated(subset=['Id', 'ActivityHour'])

# Get the duplicate rows
duplicate_rows_all = hourlyCalories[duplicates_specific]
```

```python
duplicate_rows_all
```

### Add column

```python
# Convert ActivityHour to datetime format
hourlyCalories['ActivityHour'] = pd.to_datetime(hourlyCalories['ActivityHour'])

# Extract date
hourlyCalories['Date'] = hourlyCalories['ActivityHour'].dt.date

# Extract day of the week
hourlyCalories['DayOfWeek'] = hourlyCalories['ActivityHour'].dt.day_name()

# Extract hour
hourlyCalories['Hour'] = hourlyCalories['ActivityHour'].dt.hour
```

```python
hourlyCalories
```

```python
hourlyCalories.to_csv('hourlyCalories_cleaned.csv', index=False)
```
