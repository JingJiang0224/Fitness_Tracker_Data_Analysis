### Delete the outliers in original set

```python
# dailyActivity_merged

dailyActivity_merged['TotalDistance-1'] = dailyActivity_merged['TotalDistance'].round(2)
dailyActivity_merged['SumDistance-1'] = (df['TrackerDistance'] + df['LoggedActivitiesDistance']).round(2)

dailyActivity_merged['Equal'] = dailyActivity_merged['TotalDistance-1'] == dailyActivity_merged['SumDistance-1']
dailyActivity_merged[dailyActivity_merged['Equal'] == False]
```

```python
# Condition for string match
condition = df['ActivityDate'] == '2024-06-30'

# Drop rows meeting the condition
df_cleaned = df.drop(dailyActivity_merged[dailyActivity_merged['Equal'] == False].index)
```

```python
df_cleaned
```

## Data Manipulation

### 1 - Add day week, month

```python
# Get day of the week from 'ActivityDate'

# Convert ActivityDate to datetime
df_cleaned['ActivityDate'] = pd.to_datetime(df_cleaned['ActivityDate'], format='%m/%d/%Y', errors='coerce')

# Verify conversion
if df_cleaned['ActivityDate'].isnull().any():
    print("There are some dates that could not be converted. They will be set as NaT (Not a Time).")
    print(df_cleaned[df_cleaned['ActivityDate'].isnull()])

# Extract day of the week
df_cleaned['DayOfWeek'] = df_cleaned['ActivityDate'].dt.day_name()
```

```python
# Extract month name
df_cleaned['MonthName'] = df_cleaned['ActivityDate'].dt.month_name()
```

```python
df_cleaned.head()
```

### 2 - Add additional columns

```python
# step length
df_cleaned['step_length'] = df_cleaned['TotalDistance'] / df_cleaned['TotalSteps']
```

```python
# TotalMinutes
df_cleaned['TotalMinutes'] = df_cleaned['VeryActiveMinutes'] + df_cleaned['FairlyActiveMinutes'] + df_cleaned['LightlyActiveMinutes'] + df_cleaned['SedentaryMinutes']
```

```python
# Average calories burned per active minute
df_cleaned['AvgCalories'] = df_cleaned['Calories'] / df_cleaned['TotalMinutes']
```

```python
df_cleaned.head()
```

```python
# Very Active Ratio
df_cleaned['VeryActiveRatio'] = df_cleaned['VeryActiveMinutes'] / df_cleaned['TotalMinutes']
df_cleaned['FairlyActiveRatio'] = df_cleaned['FairlyActiveMinutes'] / df_cleaned['TotalMinutes']
df_cleaned['LightlyActiveRatio'] = df_cleaned['LightlyActiveMinutes'] / df_cleaned['TotalMinutes']
df_cleaned['SedentaryRatio'] = df_cleaned['SedentaryMinutes'] / df_cleaned['TotalMinutes']

```

```python
df_cleaned.columns
```

```python
df_output = df_cleaned[['Id', 'ActivityDate', 'DayOfWeek', 'MonthName', 'TotalSteps', 'TotalDistance', 'TrackerDistance',
       'LoggedActivitiesDistance', 'VeryActiveDistance',
       'ModeratelyActiveDistance', 'LightActiveDistance',
       'SedentaryActiveDistance', 'VeryActiveMinutes', 'FairlyActiveMinutes',
       'LightlyActiveMinutes', 'SedentaryMinutes', 'Calories',
       'step_length',
       'TotalMinutes', 'AvgCalories', 'VeryActiveRatio', 'FairlyActiveRatio',
       'LightlyActiveRatio', 'SedentaryRatio']]
```

```python
df_output
```

```python
import pandas as pd
df_output = pd.read_csv('dailyActivity_merged_cleaned.csv')  


# Count the occurrences of each user (Id)
user_counts = df_output.groupby('Id').size().reset_index(name='Count')
```

```python
user_counts.head()
```

```python
user_counts.min(), user_counts.max()
```

```python
# Define bins and labels
# LightActiveUser: 0-10
# ModerateActiveUser: 11-20
# VeryActiveUser: >21
bins = [0, 10, 20, float('inf')]
labels = ['LightActiveUser', 'ModerateActiveUser', 'VeryActiveUser']

# Use pd.cut to classify
user_counts['Classification'] = pd.cut(user_counts['Count'], bins=bins, labels=labels, right=False)
```
