```python
import pandas as pd
```

```python
hourlyIntensities = pd.read_csv('hourlyIntensities_merged.csv')
```

```python
hourlyIntensities
```

### Missing data & duplicate data

```python
duplicates_specific = hourlyIntensities.duplicated(subset=['Id', 'ActivityHour'])

# Get the duplicate rows
duplicate_rows_all = hourlyIntensities[duplicates_specific]
```

```python
duplicate_rows_all
```

### Add columns

```python
# Convert ActivityHour to datetime format
hourlyIntensities['ActivityHour'] = pd.to_datetime(hourlyIntensities['ActivityHour'])

# Extract date
hourlyIntensities['Date'] = hourlyIntensities['ActivityHour'].dt.date

# Extract day of the week
hourlyIntensities['DayOfWeek'] = hourlyIntensities['ActivityHour'].dt.day_name()

# Extract hour
hourlyIntensities['Hour'] = hourlyIntensities['ActivityHour'].dt.hour
```
